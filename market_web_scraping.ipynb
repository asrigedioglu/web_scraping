
Market Web Scraping


```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
def get_carrefourSA(url):

    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text,"html.parser")

    else:
        print(f"İstek başarısız.Status Code: {response.status_code}")
       
    products = soup.find_all("div", {"class":"product_click"})

    
    carrefourSA_liste =[]
    
    for p in products:
        product_brand = p.find("input",{"id":"productBrandNamePost"}).get("value")
        product_code = p.find("input",{"id":"productCodePost"}).get("value")
        product_price = p.find("input",{"id":"productPricePost"}).get("value")
        product_name = p.find("input",{"id":"productNamePost"}).get("value")
        product_main = p.find("input",{"id":"productMainCategoryPost"}).get("value")
        product_total = p.find("input",{"id":"productTotalCategory"}).get("value")
        product_discount = p.find("input",{"id":"productDiscountPricePost"}).get("value")

        carrefourSA_liste.append({
                                 "Brands":product_brand,
                                 "Code":product_code,
                                 "Price":product_price,
                                 "Name":product_name,
                                 "Main":product_main,
                                 "Total":product_total,
                                 "Discount_Price":product_discount})
        
    return carrefourSA_liste

camasir = get_carrefourSA("https://www.carrefoursa.com/camasir-yikama-urunleri/c/1627?q=%3AbestSeller%3AinStockFlag%3Atrue&show=All")
bulasik = get_carrefourSA("https://www.carrefoursa.com/bulasik-yikama-urunleri/c/1613?q=%3AbestSeller%3AinStockFlag%3Atrue&show=All")
ev_bakim = get_carrefourSA("https://www.carrefoursa.com/ev-bakimi/c/1557?q=%3AbestSeller%3AinStockFlag%3Atrue&show=All")
```

```python
camasir_df = pd.DataFrame(camasir)
bulasik_df = pd.DataFrame(bulasik)
ev_bakim_df = pd.DataFrame(ev_bakim)

carreforSA_df = pd.concat([camasir_df,bulasik_df,ev_bakim_df])
carreforSA_df.info()
```

```python
carreforSA_df[["Price","Discount_Price"]] = carreforSA_df[["Price","Discount_Price"]].astype("float64")
carreforSA_df.info()
```

```python
carreforSA_df["Actual_Price"] = carreforSA_df["Price"] - carreforSA_df["Discount_Price"]
carreforSA_df["Market_Place"] = "CarreforSA"
```

ŞOK MARKET

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import numpy as np
```

```python
pd.options.display.max_rows=300
```

```python
def prepare_driver(path):
    # Driver_path chrome version'a uygun path eklemesini yaptık. Yer tespiti yaptık.
    driver = webdriver.Chrome()
    driver.set_window_size(1250,740)
    driver.set_window_position(0,0)
    
    return driver

def get_sok(url):
    driver = prepare_driver(r"C:\Users\Lenovo\Downloads\chromedriver-win32\chromedriver-win32\chromedriver.exe")
    driver.get(url)
    # Veri çekmek istediğiniz web sayfasının URL'sini getirir.
    
    product_information = driver.find_elements(By.CLASS_NAME,"productbox-content")
   
    product_data = []
    
    for info in product_information: 
            sep= info.text.splitlines()  #Satırları alt alta ayırmaya yarar.
            product_data.append(sep)
   
            
   
    return product_data


bulasik = get_sok("https://www.sokmarket.com.tr/bulasik-c-1442")
camasir = get_sok("https://www.sokmarket.com.tr/camasir-c-1443")
ev = get_sok("https://www.sokmarket.com.tr/ev-yuzey-temizleyicileri-c-1444")

bulasik_df = pd.DataFrame(data=bulasik,columns=["Price","Name","Old"])
camasir_df = pd.DataFrame(data=camasir,columns=["Price","Name","Old"])
ev_df = pd.DataFrame(data=ev,columns=["Price","Name","Old"])

```

```python
df_sok = pd.concat([bulasik_df,camasir_df,ev_df]);df_sok
df_sok["Market_Place"] = "Şok"
df_sok


df_sok.drop(df_sok[df_sok["Old"].notna()].index,inplace=True)
df_sok.drop('Old', axis=1, inplace=True)

```

```python
df_sok["Brands"] = ""

for index,row in df_sok.iterrows():
    brand = row["Name"].split(" ")[0]
    df_sok.at[index,"Brands"] = brand
    
# df_sok iterrows ile DataFrame üzerinde satır satır dönmek için kullanılır. Her satırda ki index'e göre veri seçilir ve
# Brands kolonuna o indexte ki brand name gelir.
```

A101 MARKET


```python
def get_a101(url):
    driver = prepare_driver(r"C:\Users\Lenovo\Downloads\chromedriver-win32\chromedriver-win32\chromedriver.exe")
    # Prepare_driver fonksiyonu daha önce Şok datasında kullandığımız fonksiyonudr.
    driver.get(url)
    # Veri çekmek istediğiniz web sayfasının URL'si
    
    product_names = driver.find_elements(By.TAG_NAME,"h3")
    product_prices = driver.find_elements(By.CLASS_NAME,"current")
    
    product_name_data = []
    price_data = []
    brand_data = []

    for product in product_names:
        product_name_data.append(product.text)
        
    for brands in product_name_data:
            sentence = brands.split(" ")[0]
            brand_data.append(sentence)
    
    for price in product_prices:
        price_data.append(price.text)
        
    return product_name_data, price_data, brand_data


product_name_data, price_data, brand_data = get_a101("https://www.a101.com.tr/market/ev-bakim-temizlik/?in_stock=1")


product_name_df = pd.DataFrame(data=product_name_data,columns=["Name"])
price_df = pd.DataFrame(data=price_data,columns=["Price"])
brand_df = pd.DataFrame(data=brand_data,columns=["Brands"])

df_a101 = pd.concat([product_name_df,price_df,brand_df],axis=1)
df_a101["Market_Place"] = "A101"
df_a101
```
